# Highest Grossing Mobile Games - EDA

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('highest_grossing_mobile_games_cleaned.csv')
df.head()
```

## Dataset Info and Cleaning

```python
df.info()
df.describe(include='all')
```

## Top 10 Grossing Games

```python
top_10 = df.sort_values(by='Revenue', ascending=False).head(10)
top_10[['Name', 'Revenue']]
```

## Revenue Distribution

```python
plt.figure(figsize=(10,6))
sns.histplot(df['Revenue'], bins=20, kde=True)
plt.title('Revenue Distribution')
plt.xlabel('Revenue (USD)')
plt.ylabel('Number of Games')
plt.show()
```

## Game Genres Distribution

```python
plt.figure(figsize=(10,6))
df['Genre'].value_counts().head(10).plot(kind='bar')
plt.title('Top 10 Game Genres')
plt.xlabel('Genre')
plt.ylabel('Number of Games')
plt.xticks(rotation=45)
plt.show()
```

## Games Released Over Time

```python
# Convert 'Year' to datetime and check for valid entries
df['Year'] = pd.to_datetime(df['Year'], errors='coerce')
valid_years = df['Year'].dropna()
if not valid_years.empty:
    valid_years.dt.year.value_counts().sort_index().plot(kind='bar', figsize=(12,6))
    plt.title('Number of Game Releases by Year')
    plt.xlabel('Year')
    plt.ylabel('Games Released')
    plt.show()
else:
    print('No valid datetime entries in Year column to plot.')
```
